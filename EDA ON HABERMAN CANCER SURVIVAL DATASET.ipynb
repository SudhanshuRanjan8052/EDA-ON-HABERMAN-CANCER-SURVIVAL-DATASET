# About the Haberman Cancer Survival Dataset:

Haberman's dataset contains data from the study conducted in chicago's Billing Hospital between year 1958 to 1970 
for the patients who undergone the surgery of breast cancer.

There are 4 attributes in the dataset out of which 3 are features and 1 is class attribute as follows:
1. Age 
2. Year of operation
3. Number of axiliary nodes(Lymph nodes)
4. survival status -> class attribute

Lymph nodes:
A person’s lymphatic system is part of their immune system. The lymphatic system is a series of connected nodes and vessels.
Lymph vessels transport a substance called lymph fluid around a person’s body. Lymph fluid contains white blood cells, which help a person to fight infections.
Lymph nodes are small structures that filter foreign substances, such as cancer cells or infections, from the lymph fluid. Lymph nodes also contain white blood cells that attack invading organisms in the lymph fluid.
When cancer cells appear in a person’s lymph nodes, it means cells have broken away from the original tumor and traveled via the lymph vessels to a lymph node.
The presence of cancer cells inside a lymph node indicates that the cancer is spreading. Cancer cells that survive inside the lymph node may then travel to other parts of the body.
Here, number of axiliary nodes indicates the number of nodes got infected by the cancer. Number is directly proportional to  the amount of spread of cancer from the original tumor, if the number is more then the cancer has been spreaded far away from the original tumor.

Age: It represent the age of patient at which they undergone surgery (age from 30 to 83).

Operation year: Year in which patient was undergone surgery(1958–1969).

Survival Status: It represent whether patient survive more than 5 years or less after undergone through surgery.Here if patients survived 5 years or more is represented as 1 and patients who survived less than 5 years is represented as 2.

In the above plot, we can observe that people having less infected axiliary nodes survives more while people with more infected axiliary nodes survives less. We can see that this plot is not enough clear to predict the survival status but still with the above observation we can conclude the following:

else if(no. of axiliary nodes >=0 and no. of axiliary nodes<=3.5) then status would be 1 i.e, long survival

else if(no. of axiliary nodes>3.5) then status would be 2 i.e, short survival

By above conclusion we can only say that the patient with a particular no. of infected lymph nodes would survive less or more but cannot say the percentage of people who will survive less or more. 

# 2. Cumulative Distribution Function (CDF):

# (i) AGE:

```python
#CDF OF STATUS 1
counts,bin_edges = np.histogram(status_1["age"],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#CDF OF STATUS 2
counts,bin_edges = np.histogram(status_2["age"],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

In the above plot, we can clearly observe that the CDF of both long survival status as well as short survival status are overlapping each other. Therefore, from the above CDF plot, we cannot predict the percentage of people who will survive more or less.

# (ii) YEAR:

```python
#CDF OF STATUS 1
counts,bin_edges = np.histogram(status_1["year"],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#CDF OF STATUS 2
counts,bin_edges = np.histogram(status_2["year"],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

In the above plot, we can clearly observe that the CDF of both long survival status as well as short survival status are overlapping each other. Therefore, from the above CDF plot, we cannot predict the percentage of people who will survive more or less.

# (iii) NODES:

```python
#CDF OF STATUS 1
counts,bin_edges = np.histogram(status_1["nodes"],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#CDF OF STATUS 2
counts,bin_edges = np.histogram(status_2["nodes"],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

In the above plot, we can observe that both the survival status not completely overlapping each other. Therefore, we can use nodes feature to calculate the percentage of survival of patient.

Firstly, if we just look at long survival status(yellow), there are 85% chance of long survival if infected nodes detected is less than 5. There is 80-85 % chance of long survival if there is nodes detected is less. As the number of infected nodes increases, the chance of long survival decreases and chances of short survival increases. 

If the number of detected nodes <5 then there is a 55% chance of short survival which increases with number of nodes increases.
When the number of nodes >40 there is a 100% of people have a chance of short survival.

# Mean, Standard deviation, Median, Percentile, Quantile, MAD:

```python
print("Means:")
print(np.mean(status_1["nodes"]))
print(np.mean(status_2["nodes"]))

print("\nStandard deviation:")
print(np.std(status_1["nodes"]))
print(np.std(status_2["nodes"]))
```

Long survival mean is 2.79 and short survival mean is 7.45. It shows that probability of short survival is more in data set.

Standard deviation of long survival is 5.85 and short survival is 9.12. It shows that data of short survival is spreaded more.

```python
print("Median:")
print(np.median(status_1["nodes"]))
print(np.median(status_2["nodes"]))

print("\nPercentiles:")
print(np.percentile(status_1["nodes"],90))
print(np.percentile(status_2["nodes"],90))

print("\nQuantiles:")
print(np.percentile(status_1["nodes"], np.arange(0,100,25)))
print(np.percentile(status_2["nodes"], np.arange(0,100,25)))
```

Observation:

From above observation it is clear that average axillary nodes in long survival is 0 and for short survival it is 4. ie, Patients who have average 4 auxillary nodes have short survival status.

Quantiles shows that nearly 50th% of axillary nodes are 0 in long survival and 75th% of patients have nodes less than 3 that is 25% patients are having nodes more than 3.
Similarly, In short survival 75th% of patients have minimum 11 nodes detected.

At 90th% there if nodes detected is >8 then it has long survival status and if nodes are >20 then patients will have short survival status

```python
from statsmodels import robust
print("Median Absolute Deviation:")
print(robust.mad(status_1["nodes"]))
print(robust.mad(status_2["nodes"]))
```
